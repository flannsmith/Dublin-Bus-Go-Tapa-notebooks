```python
official_routes = {}
for j in df['JOURNEY_PATTERN_ID'].unique():
    
    gf = df[df['JOURNEY_PATTERN_ID']==j].sort_values(by=['SEQUENCE_NUMBER'])
    official_routes[j]=[int(i) for i in gf['ID'].unique()]
import pickle
with open('officalroutes.pickle','wb') as handle:
    pickle.dump(official_routes, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
official_routes
```

```python
missing_stops = []
missing_links = []
import os
for route in official_routes:
    
    for i in range(0, len(official_routes[route])-1):
        stops =official_routes[route]
        
        if not os.path.isdir('/home/student/data/stops/'+str(stops[i])):
            missing_stops.append(stops[i])
        elif not os.path.exists('/home/student/data/stops/'+str(stops[i])+'/'+str(stops[i+1])+'.csv'):
            missing_links.append(str(stops[i])+'_'+str(stops[i+1]))
            
    
    
    
```

## Looking for the missing stops in the leavetimes files

```python
missed_stops=set(missing_stops)
```

```python
df = pd.read_csv('/home/student/data/rt_leavetimes_2016_I_DB.txt',delimiter=';',chunksize = 1000000)
found_stops = set()
while True:
    chunk = df.get_chunk(1000000)
    stops = [int(s) for s in chunk['stoppointid'].unique()]
    for stop in stops:
        if stop in missed_stops and stop not in found_stops:
            
            found_stops.add(stop)
    
```
