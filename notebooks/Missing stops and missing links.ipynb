```python
official_routes = {}
for j in df['JOURNEY_PATTERN_ID'].unique():
    
    gf = df[df['JOURNEY_PATTERN_ID']==j].sort_values(by=['SEQUENCE_NUMBER'])
    official_routes[j]=[int(i) for i in gf['ID'].unique()]
import pickle
with open('officalroutes.pickle','wb') as handle:
    pickle.dump(official_routes, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
missing_stops = set()
missing_links = set()
broken_routes = set()

reverse_paths = 0
import os
out_routes={}
for route in official_routes:
    rt=[]
    for i in range(0, len(official_routes[route])-1):
        stops =official_routes[route]
        found = True
        if not os.path.isdir('/home/student/data/stops/'+str(stops[i])):
            found = False
            missing_stops.add(stops[i])
            broken_routes.add(route)
        if not os.path.exists('/home/student/data/stops/'+str(stops[i])+'/'+str(stops[i+1])+'.csv'):
            found = False
            missing_links.add(str(stops[i])+'_'+str(stops[i+1]))
            broken_routes.add(route)
            if os.path.exists('/home/student/data/stops/'+str(stops[i+1])+'/'+str(stops[i])+'.csv'):
                reverse_paths+=1
        rt.append(found)
    out_routes[route]=rt

print(reverse_paths)
print(len(missing_stops))
print(len(missing_links))
print(len(broken_routes))
print(len(official_routes))
    
    
```

```python
for route in out_routes:
    if False in out_routes[route]:
        print(route,out_routes[route])
```

## Looking for the missing stops in the leavetimes files

```python
df = pd.read_csv('/home/student/data/rt_leavetimes_2017_I_DB.txt',delimiter=';',chunksize = 1000000)
found_stops = set()
while True:
    chunk = df.get_chunk(1000000)
    stops = [int(s) for s in chunk['stoppointid'].unique()]
    for stop in stops:
        if stop in missing_stops and stop not in found_stops:
            
            found_stops.add(stop)
    
```
