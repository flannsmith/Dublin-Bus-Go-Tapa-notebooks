```python
def trim_route(route):
    try:
        new_route = []
        for trip in route:

            trip = [trip[0]] + [int(i[-4:]) for i in trip[1:]]
            new_route.append(trip)
    except:
        print(route)
    return new_route
    
```

```python
new_route_dict = {}
for route in routes_dict:
   
    new_route_dict[route.split('-')[1]] = trim_route(routes_dict[route])

```
